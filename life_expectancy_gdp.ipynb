# Life Expectancy GDP Project


###### To explore the relationship between gross domestic product and life expectancy across 6 countries.

## 
#### Glossary:
#####  life expectancy at birth in years (LEB-Y)
##### gross domestic product (GDP) - measure of the size and health of a country's economy over a period of time.
##### GDP in trillions of dollars (GDP-T)

#### Import modules

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
```

#### Read Data

```python
life_gdp = pd.read_csv('all_data.csv')
print(life_gdp.head())
```

#### Clean Data

```python
#Convert GDP to million to make easier to read 
#Find out data type of GDP column
print(np.dtype(life_gdp['GDP']))
#Add column for GDP in trillions of dollars
life_gdp['GDP-T'] = life_gdp['GDP'] / 1000000000000
#Check new column in dataframe
print(life_gdp.head())
```

```python
#Rename Life expectancy at birth (years) to be easier to work with
life_gdp.rename(columns = {'Life expectancy at birth (years)':'LEB-Y'}, inplace = True)
#Check new column name in dataframe
print(life_gdp.head())
```

#### Understanding data

```python
#Check to see if any null values in the columns
life_gdp.isnull().sum(axis = 0)
```

```python
# Find unique values of columns

#Country
countries = life_gdp['Country'].unique()

#Year
years = life_gdp['Year'].unique()

print('Unique countries: ' , countries)
print('Unique years: ' , years)
```

## Analysing the data

```python
plt.figure(figsize= (10,8))
sns.set_palette('deep')
ax = plt.subplot()
sns.lineplot(data = life_gdp, x = 'Year', y = 'GDP-T', hue = 'Country', legend = 'auto')
ax.set_title('GDP Over Time', fontsize = 16)
ax.set_ylabel('GDP in Trillions of US Dollars', fontsize = 12)
ax.set_xlabel('Year', fontsize = 12)
plt.show()
plt.clf()
```

```python
#Split line plot out to each country 
plt.figure(figsize= (15,12))
sns.set_style('darkgrid')
plt.subplots_adjust(wspace=0.5, hspace=0.5)

ax1 = plt.subplot(2,3,1)
sns.lineplot(data = life_gdp[life_gdp['Country'] == 'Chile'], x = 'Year', y = 'GDP-T', color = 'royalblue')
ax1.set_title('Chile')
ax1.set_ylabel('GDP in Trillions of US Dollars', fontsize =12)
ax1.set_xlabel('Year', fontsize = 12)
ax1.set_xticks(years)
ax1.set_xticklabels(years , rotation='vertical')


ax2 = plt.subplot(2,3,2)
sns.lineplot(data = life_gdp[life_gdp['Country'] == 'China'], x = 'Year', y = 'GDP-T', color = 'darkorange')
ax2.set_title('China')
ax2.set_ylabel('GDP in Trillions of US Dollars', fontsize =12)
ax2.set_xlabel('Year', fontsize = 12)
ax2.set_xticks(years)
ax2.set_xticklabels(years , rotation='vertical')

ax3 = plt.subplot(2,3,3)
sns.lineplot(data = life_gdp[life_gdp['Country'] == 'Germany'], x = 'Year', y = 'GDP-T', color = 'seagreen')
ax3.set_title('Germany')
ax3.set_ylabel('GDP in Trillions of US Dollars', fontsize =12)
ax3.set_xlabel('Year', fontsize = 12)
ax3.set_xticks(years)
ax3.set_xticklabels(years , rotation='vertical')

ax4 = plt.subplot(2,3,4)
sns.lineplot(data = life_gdp[life_gdp['Country'] == 'Mexico'], x = 'Year', y = 'GDP-T', color = 'firebrick')
ax4.set_title('Mexico')
ax4.set_ylabel('GDP in Trillions of US Dollars', fontsize =12)
ax4.set_xlabel('Year', fontsize = 12)
ax4.set_xticks(years)
ax4.set_xticklabels(years , rotation='vertical')

ax5 = plt.subplot(2,3,5)
sns.lineplot(data = life_gdp[life_gdp['Country'] == 'United States of America'], x = 'Year', y = 'GDP-T', color = 'rebeccapurple')
ax5.set_title('United States of America')
ax5.set_ylabel('GDP in Trillions of US Dollars', fontsize =12)
ax5.set_xlabel('Year', fontsize = 12)
ax5.set_xticks(years)
ax5.set_xticklabels(years , rotation='vertical')

ax6 = plt.subplot(2,3,6)
sns.lineplot(data = life_gdp[life_gdp['Country'] == 'Zimbabwe'], x = 'Year', y = 'GDP-T', color = 'peru')
ax6.set_title('Zimbabwe')
ax6.set_ylabel('GDP in Trillions of US Dollars', fontsize =12)
ax6.set_xlabel('Year', fontsize = 12)
ax6.set_xticks(years)
ax6.set_xticklabels(years , rotation='vertical')

plt.show()
plt.clf()
```

```python
#Bar chart to compare GDP across countries
gdp_order =life_gdp.sort_values('GDP-T',ascending = False)
color_gdp_order = ['rebeccapurple','darkorange', 'seagreen', 'firebrick', 'royalblue' , 'peru']
plt.figure(figsize = (10,5))
ax7 = plt.subplot()
ax7.set_title('Average GDP by country')
sns.barplot(data = gdp_order, x = 'Country', y = 'GDP-T', palette = color_gdp_order) 
ax7.set_ylabel('GDP in Trillions of US Dollars')
plt.show()
plt.clf()
```
